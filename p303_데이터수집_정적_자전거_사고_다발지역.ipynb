### response.json()

```python
# item 접근
w_data = response.json()['items']['item']
print(w_data)

# 필요한 item 출력
for row in range(len(w_data)) :
  print('spot_nm:', w_data[row]['spot_nm'])
  print('경도:', w_data[row]['lo_crd'])
  print('위도:', w_data[row]['la_crd'])

totalCount = response.json()['totalCount']
print('totalCount:', totalCount)
```

```python
# 데이터 저장을 하겠다 싶으면 별도로 데이터를 저장할 필요가 있다.
result = []
data = response.json()
w_data = data['items']['item']
for row in range(len(w_data)) :
  line={}
  line['spot_nm:'] = w_data[row]['spot_nm']
  line['lo_crd:'] = w_data[row]['lo_crd']
  line['la_crd:'] = w_data[row]['la_crd']
  result.append(line)

print(line)
```

```python
result
```

```python
# indent ="\t" : 들여쓰기를 탭으로 설정
# ensure_ascii = False : 유니코드로 변환되는 작업 장지

with open('./cycle.json','w',newline='',encoding ='utf-8') as f:
  json.dump(result, f, indent="\t", ensure_ascii = False) #  indent="\t", ensure_ascii = False는 키워드 순서가 바뀌어도 상관 없음.
```

```python
json_str="""
  {
    "name" : "홍길동",
    "age" : "30"
  }
"""
```

```python
type(json_str) # 데이터를 묶어서 str 유형이고
```

json.load : 파일  읽기  
json.dump : 파일 저장  
json.loads : 문자열(str) -> 객체(list, dict, tuple...)  
json.dumps: 객체(list, dict, tuple...) -> 문자열(str)  

```python
import json
# str -> dict, 문자열을 dict이라는 객체로 변환해줌
data = json.loads(json_str)
data
```

```python
type(data) # 문자를 dict 형태로 묶어서 내보냄
```

```python
# dict -> str
sn = json.dumps(data, ensure_ascii=False)
sn # 문자열 형태로 가지고 온다.
```

```python
type(sn)
```

```python
json_str2='[1,2,3,4]'
# str -> list , 문자열이 list 객체로 바뀜
data = json.loads(json_str2)
print(data)
print(type(data))
```
